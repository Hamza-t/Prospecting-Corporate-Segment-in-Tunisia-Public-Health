```python
#function
#from one page 
def getDentistsInfos_FromOnePage(url):
  result = requests.get(url).text
  doc = BeautifulSoup(result, "html.parser")
  dentists = doc.find_all(class_="col-lg-6 col-md-6 col-sm-6 col-xs-12")

  dentistsList = []
  for dentist in dentists:
    link = dentist.find("a").get("href")
    requests.get(link).text
    response = requests.get(link).text
    soup = BeautifulSoup(response, "html.parser")
    if soup.find(class_="col-lg-5 col-md-5 col-sm-5 col-xs-12"):  
      infos = soup.find(class_="col-lg-5 col-md-5 col-sm-5 col-xs-12")
      dentistInfos = {}
      if infos.find(itemprop="name"):
        dentistInfos["name"] = ((infos.find(itemprop="name").text).split('-'))[0]
      if infos.find(itemprop="streetAddress"):
        dentistInfos["streetAddress"] = infos.find(itemprop="streetAddress").text
      if infos.find(itemprop="addressLocality"):
        dentistInfos["addressLocality"] = infos.find(itemprop="addressLocality").text
      if infos.find(itemprop="streetAddress"):
        dentistInfos["MapsLink"] = infos.find("a", itemprop="streetAddress").get("href")
      if infos.find_all(itemprop="telephone"):
        dentistInfos["telephone"] = []
        for telephone in infos.find_all(itemprop="telephone"):
          dentistInfos["telephone"].append(((telephone.text).split(":"))[1])
      dentistsList.append(dentistInfos)

  return dentistsList
```

```python
InfosFromOnePage = pd.DataFrame(getDentistsInfos_FromOnePage(url))
InfosFromOnePage
```

```python
#From all the pages 
link = "https://tunisie-medicale.com/index.php/dentiste/index/"
InfosFromAllPages = pd.DataFrame()
for i in range(0, 144):
  InfosFromAllPages = pd.concat([InfosFromAllPages, pd.DataFrame(getDentistsInfos_FromOnePage(link + str(i*20)))])
  print("page "+ str(i+1) +" done")
InfosFromAllPages
```

```python
InfosFromAllPages = InfosFromAllPages.reset_index()
InfosFromAllPages
```
