```python
#data cleaning 
data = data[["name","streetAddress","addressLocality","telephone"]]
```

```python
data.head()
```

```python
#Cleaning 
#ville
def ville(t):
  t = str(t)
  return " ".join(t.split(" ")[1:])

#gouvernorat
def gouvernorat(t):
  t = str(t)
  return t.split(" ")[0]

#telephone
def phone1(t):
  t=str(t)
  tr = t.split("'")
  if len(tr)>1: 
    return tr[1][1:]
  else : return 0


def phone2(t):
  t=str(t)
  tr = t.split("'")
  if len(tr)>3: 
    return tr[3][1:]
  else : return 0


data["Cite"] = data["addressLocality"].apply(ville)
data["gouvernoratz"] = data["addressLocality"].apply(gouvernorat)
data["phone 1"] = data["telephone"].apply(phone1)
data["phone 2"] = data["telephone"].apply(phone2)
```
