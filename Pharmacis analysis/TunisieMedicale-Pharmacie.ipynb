```python
#function
#from one page 
def getPharmaciesInfos_FromOnePage(url):
  result = requests.get(url).text
  doc = BeautifulSoup(result, "html.parser")
  pharmacies = doc.find_all(class_="col-lg-6 col-md-6 col-sm-6 col-xs-12")

  pharmaciesList = []
  for pharmacy in pharmacies:
    link = pharmacy.find("a").get("href")
    requests.get(link).text
    response = requests.get(link).text
    soup = BeautifulSoup(response, "html.parser")
    if soup.find(class_="col-lg-5 col-md-5 col-sm-5 col-xs-12"):  
      infos = soup.find(class_="col-lg-5 col-md-5 col-sm-5 col-xs-12")
      pharmacyInfos = {}
      if infos.find(itemprop="name"):
        pharmacyInfos["name"] = ((infos.find(itemprop="name").text).split('-'))[0]
      if infos.find("h3"):
        Day_Night = (infos.find("h3").text)
        if "Pharmacie de jours" in Day_Night:
          pharmacyInfos["Day_Night"] = "Pharmacie de jours"
        if "Pharmacie de nuit" in Day_Night:
          pharmacyInfos["Day_Night"] = "Pharmacie de nuit"
      if infos.find(itemprop="address"):
        pharmacyInfos["address"] = (infos.find(itemprop="address")).find("a").text
      if infos.find(itemprop="addressLocality"):
        pharmacyInfos["addressLocality"] = infos.find(itemprop="addressLocality").text
      if infos.find(itemprop="address"):
        pharmacyInfos["MapsLink"] = (infos.find(itemprop="address")).find("a").get("href")
      if infos.find_all(itemprop="telephone"):
        pharmacyInfos["telephone"] = []
        for telephone in infos.find_all(itemprop="telephone"):
          pharmacyInfos["telephone"].append(((telephone.text).split(":"))[1])
      pharmaciesList.append(pharmacyInfos)

  return pharmaciesList
```

```python
InfosFromOnePage = pd.DataFrame(getPharmaciesInfos_FromOnePage(url))
InfosFromOnePage
```

```python
#From all the pages 
link = "https://tunisie-medicale.com/index.php/pharmacie/index/"
InfosFromAllPages = pd.DataFrame()
for i in range(0, 81):
  InfosFromAllPages = pd.concat([InfosFromAllPages, pd.DataFrame(getPharmaciesInfos_FromOnePage(link + str(i*20)))])
  print("page "+ str(i+1) +" done")
InfosFromAllPages
```

```python
InfosFromAllPages = InfosFromAllPages.reset_index()
InfosFromAllPages
```
