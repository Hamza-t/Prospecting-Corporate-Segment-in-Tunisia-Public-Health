```python
#data cleaning 
data = data[["name","speciality","adress","local","telephones"]]
```

```python
data.head()
```

```python
#Cleaning 
#ville
def ville(t):
  t = str(t)
  return " ".join(t.split(" ")[1:])

#gouvernorat
def gouvernorat(t):
  t = str(t)
  return t.split(" ")[0]

#telephone
def phone1(t):
  t=str(t)
  tr = t.split("'")
  if len(tr)>1: 
    return tr[1][4:]
  else : return 0


def phone2(t):
  t=str(t)
  tr = t.split("'")
  if len(tr)>3: 
    return tr[3][4:]
  else : return 0




t= data["telephones"][0]
t.split("'")[3][4:]

data["ville"] = data["local"].apply(ville)
data["gouvernorat"] = data["local"].apply(gouvernorat)
data["phone 1"] = data["telephones"].apply(phone1)
data["phone 2"] = data["telephones"].apply(phone2)
```
